```python
def imputation(tsdata_valuelist, seasonality, impute_thres=1.96):
    def seasonalDiff(tsdata_valuelist, seasonality):
        tsdata_valuelist[np.isnan(tsdata_valuelist)] = tsdata_valuelist.rolling(
            window=int(len(tsdata_valuelist) / 4), min_periods=0, center=True
        ).mean()
        back_Diff = np.diff(tsdata_valuelist)
        front_Diff = (
            tsdata_valuelist[:1].values[0] - tsdata_valuelist[::-1][:1].values[0]
        )
        diff_data = np.append(front_Diff, back_Diff) #append해서 순서대로 보여줌

        if seasonality > 1 or seasonality is not None:
            front_Diff_seasonal = (
                diff_data[:seasonality] - diff_data[::-1][:seasonality][::-1]
            )
            back_Diff_seasonal = (
                diff_data[seasonality : (len(diff_data))]
                - diff_data[0 : (len(diff_data) - seasonality)]
            )
            diff_data = np.append(front_Diff_seasonal, back_Diff_seasonal)
        return diff_data

    try:
        impute_tsdata = copy.deepcopy(tsdata_valuelist)
        diff_data = seasonalDiff(impute_tsdata, seasonality)
        NaFilter_data = diff_data[~np.isnan(diff_data)]
        diff_std = np.std(NaFilter_data)

        # extract train anomaly
        inds = list(
            np.any(
                [
                    diff_data < (-diff_std * impute_thres),
                    diff_data > (diff_std * impute_thres),
                ],
                axis=0,
            )
        )
        impute_tsdata[inds] = np.nan

        # first impute
        windowSmoothing = impute_tsdata.rolling(
            window=int(len(impute_tsdata) / 4), min_periods=0, center=True
        ).mean()

        # doubling impute
        if np.sum(np.isnan(windowSmoothing).astype(int)) != 0:
            windowSmoothing = (
                windowSmoothing.rolling(
                    window=int(len(impute_tsdata) / 2), min_periods=0, center=True
                )
                .mean()
                .fillna(0)
            )
        impute_tsdata[np.isnan(impute_tsdata)] = windowSmoothing[
            impute_tsdata[np.isnan(impute_tsdata)].keys()
        ]
        return impute_tsdata
    except Exception as ex:
        print(ex)
        return "(imputaion) imputation error"
```
